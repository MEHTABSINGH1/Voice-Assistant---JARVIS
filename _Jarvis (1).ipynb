```python
 
import pyttsx3
import datetime
import speech_recognition as sr
import wikipedia
import webbrowser as wb
import os
import random
import pyautogui
import tkinter as tk
from tkinter import ttk, Text

engine = pyttsx3.init()

def speak(audio):
    engine.say(audio)
    engine.runAndWait()

def get_time():
    Time = datetime.datetime.now().strftime("%I:%M:%S")
    speak("The current time is")
    speak(Time)
    output_text.insert(tk.END, f"The current time is {Time}\n")

def get_date():
    day = int(datetime.datetime.now().day)
    month = int(datetime.datetime.now().month)
    year = int(datetime.datetime.now().year)
    speak("The current date is")
    speak(day)
    speak(month)
    speak(year)
    output_text.insert(tk.END, f"The current date is {day}/{month}/{year}\n")

def wish_me():
    output_text.insert(tk.END, "Welcome back Mehtab Singh!!\n")
    speak("Welcome back Mehtab Singh!!")
    
    hour = datetime.datetime.now().hour
    if hour >= 4 and hour < 12:
        speak("Good Morning Mehtab Singh!!")
        output_text.insert(tk.END, "Good Morning Mehtab Singh!!\n")
    elif hour >= 12 and hour < 16:
        speak("Good Afternoon Mehtab Singh!!")
        output_text.insert(tk.END, "Good Afternoon Mehtab Singh!!\n")
    elif hour >= 16 and hour < 24:
        speak("Good Evening Mehtab Singh!!")
        output_text.insert(tk.END, "Good Evening Mehtab Singh!!\n")
    else:
        speak("Good Night Mehtab Singh, See You Tomorrow")

    speak("Jarvis at your service Mehtab Singh, please tell me how may I help you.")
    output_text.insert(tk.END, "Jarvis at your service Mehtab Singh, please tell me how may I help you.\n")

def take_screenshot():
    img = pyautogui.screenshot()
    img_path = os.path.expanduser("~\\Pictures\\ss.png")
    img.save(img_path)
    speak("I've taken a screenshot, please check it.")
    output_text.insert(tk.END, "Screenshot taken.\n")

def take_command():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        speak("Listening...")
        output_text.insert(tk.END, "Listening...\n")
        r.pause_threshold = 1
        audio = r.listen(source)

    try:
        speak("Recognizing...")
        output_text.insert(tk.END, "Recognizing...\n")
        query = r.recognize_google(audio, language="en-in")
        output_text.insert(tk.END, f"You said: {query}\n")
    except Exception as e:
        output_text.insert(tk.END, f"Error: {e}\n")
        speak("Please say that again")
        return "Try Again"

    return query.lower()

def execute_command(query):
    if "time" in query:
        get_time()
    elif "date" in query:
        get_date()
    elif "who are you" in query:
        speak("I'm JARVIS created by Mr. Kishan and I'm a desktop voice assistant.")
        output_text.insert(tk.END, "I'm JARVIS created by Mr. Kishan and I'm a desktop voice assistant.\n")
    elif "how are you" in query:
        speak("I'm fine Mehtab Singh, What about you?")
        output_text.insert(tk.END, "I'm fine Mehtab Singh, What about you?\n")
    elif "fine" in query or "good" in query:
        speak("Glad to hear that Mehtab Singh!!")
        output_text.insert(tk.END, "Glad to hear that Mehtab Singh!!\n")
    elif "wikipedia" in query:
        try:
            speak("Ok wait Mehtab Singh, I'm searching...")
            query = query.replace("wikipedia", "")
            result = wikipedia.summary(query, sentences=2)
            speak(result)
            output_text.insert(tk.END, f"{result}\n")
        except:
            speak("Can't find this page Mehtab Singh, please ask something else")
            output_text.insert(tk.END, "Can't find this page Mehtab Singh, please ask something else\n")
    elif "open youtube" in query:
        wb.open("https://www.youtube.com")
    elif "open google" in query:
        wb.open("https://www.google.com")
    elif "open stack overflow" in query:
        wb.open("https://stackoverflow.com")
    elif "play music" in query:
        song_dir = os.path.expanduser("~\\Music")
        songs = os.listdir(song_dir)
        x = len(songs)
        y = random.randint(0, x-1)
        os.startfile(os.path.join(song_dir, songs[y]))
    elif "open chrome" in query:
        chromePath = "C:\\Program Files (x86)\\Google\\Chrome\\Application\\chrome.exe"
        os.startfile(chromePath)
    elif "search on chrome" in query:
        try:
            speak("What should I search?")
            search = take_command()
            chromePath = "C:\\Program Files (x86)\\Google\\Chrome\\Application\\chrome.exe"
            wb.get(chromePath).open_new_tab(search)
            output_text.insert(tk.END, f"Searching: {search}\n")
        except Exception as e:
            speak("Can't open now, please try again later.")
            output_text.insert(tk.END, "Can't open now, please try again later.\n")
    elif "remember that" in query:
        speak("What should I remember")
        data = take_command()
        speak(f"You said me to remember that {data}")
        output_text.insert(tk.END, f"You said me to remember that {data}\n")
        with open("data.txt", "w") as remember:
            remember.write(data)
    elif "do you remember anything" in query:
        try:
            with open("data.txt", "r") as remember:
                data = remember.read()
                speak(f"You told me to remember that {data}")
                output_text.insert(tk.END, f"You told me to remember that {data}\n")
        except FileNotFoundError:
            speak("I don't remember anything.")
            output_text.insert(tk.END, "I don't remember anything.\n")
    elif "screenshot" in query:
        take_screenshot()
    elif "weather" in query:
        get_weather()
    elif "news" in query:
        get_news()
    elif "joke" in query:
        tell_joke()
    elif "offline" in query:
        root.destroy()

def get_weather():
    api_key = "YOUR_API_KEY"
    base_url = "http://api.openweathermap.org/data/2.5/weather?"
    speak("Please tell me the city name")
    city_name = take_command()
    complete_url = base_url + "appid=" + api_key + "&q=" + city_name
    response = requests.get(complete_url)
    x = response.json()
    if x["cod"] != "404":
        y = x["main"]
        current_temperature = y["temp"]
        current_pressure = y["pressure"]
        current_humidity = y["humidity"]
        z = x["weather"]
        weather_description = z[0]["description"]
        weather_report = (f"Temperature: {current_temperature}°K\n"
                          f"Pressure: {current_pressure} hPa\n"
                          f"Humidity: {current_humidity}%\n"
                          f"Description: {weather_description}")
        speak(f"The weather in {city_name} is as follows")
        speak(weather_report)
        output_text.insert(tk.END, f"Weather in {city_name}:\n{weather_report}\n")
    else:
        speak("City Not Found")
        output_text.insert(tk.END, "City Not Found\n")

def get_news():
    api_key = "YOUR_API_KEY"
    url = f"http://newsapi.org/v2/top-headlines?country=in&apiKey={api_key}"
    response = requests.get(url)
    news = response.json()
    articles = news["articles"]
    speak("Here are the top news headlines")
    for i, article in enumerate(articles[:5], 1):
        news_headline = article["title"]
        speak(f"News {i}: {news_headline}")
        output_text.insert(tk.END, f"News {i}: {news_headline}\n")

def tell_joke():
    jokes = [
        "Why don’t scientists trust atoms? Because they make up everything!",
        "Why did the chicken join a band? Because it had the drumsticks!",
        "What do you get when you cross a snowman and a vampire? Frostbite!",
        "Why don’t skeletons fight each other? They don’t have the guts!"
    ]
    joke = random.choice(jokes)
    speak(joke)
    output_text.insert(tk.END, f"Joke: {joke}\n")

root = tk.Tk()
root.title("JARVIS Voice Assistant")
root.configure(bg="#2c3e50")

style = ttk.Style()
style.theme_use('clam')

frame = ttk.Frame(root, padding="10", style="TFrame")
frame.pack(pady=10)

output_text = Text(frame, height=20, width=60, wrap='word', bg="#ecf0f1", fg="#2c3e50",

 font=("Helvetica", 12), padx=10, pady=

10)
output_text.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

wish_me()

button_frame = ttk.Frame(root, style="TFrame")
button_frame.pack(pady=10)

listen_button = ttk.Button(button_frame, text="Start Listening", command=lambda: execute_command(take_command()), style="TButton")
listen_button.grid(row=0, column=0, padx=5)

style.configure('TFrame', background="#2c3e50")
style.configure('TButton', background="#3498db", foreground="#ffffff", font=("Helvetica", 12, 'bold'))
style.map('TButton', background=[('active', '#2980b9')])
def take_command():
    r = sr.Recognizer()

    with sr.Microphone() as source:
        output_text.insert(tk.END, "Listening...\n")
        root.update()   # refresh GUI
        
        r.adjust_for_ambient_noise(source)
        audio = r.listen(source, timeout=5, phrase_time_limit=7)

    try:
        output_text.insert(tk.END, "Recognizing...\n")
        root.update()

        query = r.recognize_google(audio, language="en-in")

        output_text.insert(tk.END, f"You said: {query}\n")

    except Exception as e:
        output_text.insert(tk.END, "Could not understand\n")
        return "none"

    return query.lower()
root.mainloop()
 
```
